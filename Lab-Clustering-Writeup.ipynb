### Overview 
For this project, I used kmeans clustering to group nba players into performance based groups, and then evaluated players that a GM could bring in to compete in the playoffs without spending too much money for big names or stars.

### Data Cleaning
The first step was data cleaning. The original stats dataframe had way too many features that didn't contribute to the analysis. I trimmed it by about two thirds, keeping only relevant identifiers and meanigful stats correlated directly to performace. I then used idmax, a command I'd never used before, to drop duplicates for players who were traded while keeping their overall stats for the year. I then merged the stats df with the salary df, and dropped any missing values.

### Model Creation
The next step was to initialize and run the kmeans model with an intial guess about the best number of clusters (I chose 4). I ran the model, and added a column to the large dataset with the cluster assigned to each player.

### Cluster Visualization
I then visualized the clusters using a 3d scatter plot mapping total games played, points, and rebounds, as I thought these were variables that would correspond most directly with salary. I also set salary as a color spectrum and cluster as shapes on the plot, which allowed us to visualize trends. The dataset was pretty continuous so the visualization wasn't very helpful in understanding the clusters, but there was an overall correlation between the stats, clusters, and salary


![](sc6.png)

### Evaluate Cluster Structure
Next, I used I examined the spread/structure of the clusters via total variance explained and silhouette scores. 

Total variance explained is 1 minus the sum of the spreads of each cluster divided by the overall spread of the data, which gives you a proportion that represents how much of the spread is explained by the clusters. Values close to 1 mean the error within the clusters is small, representing tight clusters.

Silhouette scores measure how far points are from their cluster center to other cluster centroids. Scores close to one represent tight clusters.

My model with 4 clusters had a TVE of 0.65 and a silhouette score of 0.28, neither of which are great

### Optimize k value
I then ran a for loop to run the model under different k values to compare silhouette scores and total variance across different amounts of clusters, and double checked the result with an elbow plot. The best balance between the two metrics was at k = 3. I then remade the 3d visualization with 3 clusters to see if that made it easier to interpret.

### Player Selection
The final step was then to come up with 3-4 players in 3 different tiers of fit: bad, good, and best.

First, I examined the cluster centroids to get an idea of the type of player in each. Cluster C had the best players statistically, so I chose it for my analysis. 

Next, I made a new df called candidates, keeping only players in cluster C who had higher relative performance in all of their scaled stats than their scaled salary. For example, if a player was in the candidates df and had a scaled salary of 0.35, they were in the C cluster (The best statistical players) and had a scaled relative value higher than 0.35 in every included performance metric (PTS, ASTS, RB, etc). I then sorted them by the mean of their scaled stats in descending order to put the highest performing players at the top. That sorting metric weighted every stat equally, which isn't ideal, but better than not sorting them at all.

Finally I selected 3 players for each of the 3 tiers required: bad, good, best. It is important to note that the players in the "bad" category are still great pickups, as they drastically outperform their salary, and perform at an extremely high level, even amongst NBA players. They are merely the worst of the best, but are still very solid options.
